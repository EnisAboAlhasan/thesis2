## Understanding smoothing

```python
#to understand smoothing we will create a repeating series that starts from 0 to 9 and repeats until it reaches the length of 100
# we do this to create a sort of wave in the data which will be drastic and try to see of we can smooth it
def make_rough_column(length=100):
    result = []
    chunk = list(range(length // 10))
    for _ in range(length // len(chunk)):
        result += chunk
    return pd.Series(result)
```

```python
#we create a variable called new and we store it the data that we recieve from the function that we created and then print the first 40 rows 
new = make_rough_column(length=100)
new.head(40)
```

```python
# we try smoothing by using the average of nearest neighbor, meaning x representing position ((x-1)+(x)+(x+1)/3)
def smoothing_nearest_neighbor(column):
    # Basic smoothing (nearest neighbor):
    # ===================================
    #
    # Replaces each element in the sequence with the mean of that
    # element, the previous element and the next element.
    #
    # Suppose you have this sequence in "input" column:
    # .___________________________.
    # |         |                 |
    # | column  | result          |
    # +---------+-----------------+
    # | A (50)  | (A+A+B)/3 (67)  |
    # | B (100) | (A+B+C)/3 (92)  |
    # | C (125) | (B+C+D)/3 (108) |
    # | D (100) | (C+D+E)/3 (175) |
    # | E (300) | (D+E+E)/3 (233) |
    # +---------+-----------------+
    # |   (675) |           (675) |
    # +---------+-----------------+
    #
    # Note that some values might be lost due to rounding.
    after = column.shift(1).fillna(column.iloc[0])
    before = column.shift(-1).fillna(column.iloc[-1])
    return (column + after + before) / 3
```

```python
#we try smoothing using the nearest 4 neighbors, meaning 2 position before and after, an average of total of 5 numbers
def smoothing_4nearest_neighbor(column):

    after_2 = column.shift(2).fillna(column.iloc[0])
    after = column.shift(1).fillna(column.iloc[0])
    before = column.shift(-1).fillna(column.iloc[-1])
    before_2 = column.shift(-2).fillna(column.iloc[-1])
    return (column + after + before + after_2 + before_2) / 5
```

```python
#we try smoothing using weekly meaning 3 posintion before and 3 position after resulting in 7 days
def weekly_smoothing_nearest_neighbor(column):
    after_3 = column.shift(3).fillna(column.iloc[0])
    after_2 = column.shift(2).fillna(column.iloc[0])
    after = column.shift(1).fillna(column.iloc[0])
    before = column.shift(-1).fillna(column.iloc[-1])
    before_2 = column.shift(-2).fillna(column.iloc[-1])
    before_3 = column.shift(-3).fillna(column.iloc[-1])
    return (column + after + before + after_2 + before_2 + after_3 + before_3) / 7
```

```python
#we create a variable called new smooth that uses the function nearest neighbor
new_smooth = smoothing_nearest_neighbor(new)
#new_smooth.head(20)

#we create a variable called new 4smooth that uses the function nearest 4 neighbors
new_4smooth = smoothing_4nearest_neighbor(new)
#we display the first 20 rows
new_4smooth.head(20)
```

```python
#we create a new data freame with our original series and add 2 columns, in the second column we store the variables from new_smoothed, and in the third column we store the variables from new_4smooth
new_df = pd.DataFrame(new)
new_df['smoothed'] = new_smooth
new_df['4smoothed'] = new_4smooth
new_df.head(20)
```

```python
# we create a plot to visualize the differance between the original, and smoothed data
fig, ax = plt.subplots(figsize=(20, 12))


ax.plot(new_df.index,
       new_df[0],
       label='original',
       color='blue')

ax.plot(new_df.index,
       new_df['smoothed'],
       label='smoothed',
       color='red')

ax.plot(new_df.index,
       new_df['4smoothed'],
       label='4 smoothed',
       color='purple')


# Set title and labels for axes
ax.set(xlabel="Index column",
       ylabel="Arbitrary numbers in a sequence",
       title="Smoothing from blue to red")

plt.xticks(rotation=45)
plt.legend(loc='upper left')
```
